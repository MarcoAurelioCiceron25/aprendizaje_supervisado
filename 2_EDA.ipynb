```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats


```

```python
df = pd.read_csv('./data/titanic_clean.csv')
```

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribuci√≥n de edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Sobrevivientes por clase de pasajero')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Sobrevivientes por g√©nero')
plt.xlabel('G√©nero')
plt.ylabel('Cantidad')
plt.show()
```

```python
sns.countplot(data=df, x='Embarked', hue='Survived')
plt.title('Sobrevivientes por puerto de embarque')
plt.xlabel('Puerto')
plt.ylabel('Cantidad')
plt.show()
```

```python
def categorizar_edad(age):
    if age < 2: return 'Infante'
    elif age < 10: return 'Ni√±o'
    elif age < 18: return 'Adolescente'
    elif age < 30: return 'Joven Adulto'
    elif age < 60: return 'Adulto'
    else: return 'Mayor'

df['GrupoEdad'] = df['Age'].apply(categorizar_edad)
sns.barplot(data=df, x='GrupoEdad', y='Survived')
plt.title('Tasa de supervivencia por grupo de edad')
plt.ylabel('Tasa de supervivencia')
plt.show()
```

```python
sns.countplot(data=df, x='Pclass')
plt.title('Conteo de pasajeros por clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```

```python
sns.countplot(data=df, x='Pclass', hue='Sex')
plt.title('Distribuci√≥n de g√©nero por clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```

```python
sns.countplot(data=df, x='Embarked', hue='Pclass')
plt.title('Distribuci√≥n de clase por puerto de embarque')
plt.xlabel('Puerto')
plt.ylabel('Cantidad')
plt.show()
```

```python
print("Edad m√≠nima:", df['Age'].min())
print("Edad m√°xima:", df['Age'].max())
```

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.axvline(df['Age'].mean(), color='red', linestyle='--', label='Media')
plt.axvline(df['Age'].median(), color='green', linestyle='--', label='Mediana')
plt.axvline(df['Age'].mode()[0], color='blue', linestyle='--', label='Moda')
plt.title('Distribuci√≥n de edades con media, mediana y moda')
plt.legend()
plt.show()
```

```python
sns.boxplot(x=df['Age'])
plt.title('Boxplot de edad')
plt.show()
```

```python
print("Tarifa m√≠nima:", df['Fare'].min())
print("Tarifa m√°xima:", df['Fare'].max())
```

```python
sns.histplot(df['Fare'], bins=30, kde=True)
plt.axvline(df['Fare'].mean(), color='red', linestyle='--', label='Media')
plt.axvline(df['Fare'].median(), color='green', linestyle='--', label='Mediana')
plt.axvline(df['Fare'].mode()[0], color='blue', linestyle='--', label='Moda')
plt.title('Distribuci√≥n de tarifas con media, mediana y moda')
plt.legend()
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title('Boxplot de tarifa')
plt.show()
```
